```python
import cv2
import mediapipe as mp
```

```python
mp_hands=mp.solutions.hands
mp_drawings=mp.solutions.drawing_utils
data=mp_hands.Hands(max_num_hands=1)
data
```

```python
video=cv2.VideoCapture(0)
while True:
    success,image=video.read()
    image=cv2.flip(image,1)
    color_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    result=data.process(color_image)
    tip_id=[4,8,12,16,20]
    finger_count1=0
    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            lmlst=[]
            for id,lm in enumerate(x.landmark):
                # print(id,lm.x,lm.y)
                cx=lm.x
                cy=lm.y
                lmlst.append([id,cx,cy])
                
            if len(lmlst)!=0 and len(lmlst)==21:
                finger_count=[]
                if lmlst[12][1]<lmlst[20][1]:
                    if lmlst[4][1]>lmlst[3][1]:
                        finger_count.append(0)
                    else:
                        finger_count.append(1)
                else:
                    if lmlst[4][1]<lmlst[3][1]:
                        finger_count.append(0)
                    else:
                        finger_count.append(1)

                    # all finger except thumb
                for i in range(1,5): #i=1
                    if lmlst[tip_id[i]][2]>lmlst[tip_id[i]-2][2]:
                        finger_count.append(0)
                    else:
                        finger_count.append(1)
                
                # finger_count1=finger_count.count(1)

                        # print on screen
            mp_drawings.draw_landmarks(image,x,mp_hands.HAND_CONNECTIONS)
        
        thumbs_tip_y=lmlst[4][2]
        thumbs_base_y=lmlst[2][2]
        if finger_count==[0,0,0,0,0]:
            cv2.putText(image,"Closed:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
        elif finger_count[1:]==[0,0,0,0]:
            if thumbs_tip_y<thumbs_base_y:
                cv2.putText(image,"Thumbs Up:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
            elif thumbs_tip_y>thumbs_base_y:
                cv2.putText(image,"Thumbs Down:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
        elif finger_count==[0,1,0,0,0]:
            cv2.putText(image,"Pointing:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
        elif finger_count==[0,1,1,0,0]:
            cv2.putText(image,"Victory:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
        elif finger_count==[1,1,0,0,1]:
            cv2.putText(image,"Rock:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
        elif finger_count==[1,1,1,1,1]:
            cv2.putText(image,"Opened:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
        elif finger_count==[1,1,1,0,0]:
            cv2.putText(image,"Gun:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
        elif finger_count==[1,0,0,0,1]:
            cv2.putText(image,"Call me:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
        elif finger_count==[0,0,1,1,1]:
            cv2.putText(image,"Super:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
        elif finger_count==[0,0,1,0,0]:
            cv2.putText(image,"Reported Unusual Behaviour:"+str(finger_count),(55,150),cv2.FONT_HERSHEY_COMPLEX,1.4,color=(0,0,255),thickness=3)
        
                    # print(id,lm)

        # print(lmlst)
    cv2.imshow('Image',image)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
