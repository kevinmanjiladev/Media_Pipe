```python
import cv2
import mediapipe as mp
```

```python
# Handlamarks detection
# ------------------------

mphands=mp.solutions.hands #hand landmarks
# solutions is module which contains all landmarks
mp_drawing=mp.solutions.drawing_utils
```

```python
# import Hands() class from submodule hands

data=mphands.Hands()
print(data)
```

```python
# web cam open code

video=cv2.VideoCapture(0)
while True:
    success,image=video.read()
    img1=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    result=data.process(img1)
    if result.multi_hand_landmarks:
        print(result.multi_hand_landmarks)
        for x in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(image,x,mphands.HAND_CONNECTIONS,mp_drawing.DrawingSpec(color=(255,0,0),thickness=5,circle_radius=3),mp_drawing.DrawingSpec(color=(0,0,255),thickness=5))
    cv2.imshow('Image',image)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
