```python
import cv2
import mediapipe as mp
```

```python
mp_hands=mp.solutions.hands
mp_drawings=mp.solutions.drawing_utils
data=mp_hands.Hands(max_num_hands=1)
print(data)
```

```python
video=cv2.VideoCapture(0)
while True:
    success,image=video.read()
    color_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    result=data.process(color_image)
    tip_id=[4,8,12,16,20]
    lmlst=[]
    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            for id,lm in enumerate(x.landmark):
                print(id,lm.x,lm.y)
                cx=lm.x
                cy=lm.y
                lmlst.append([id,cx,cy])
                
                if len(lmlst)!=0 and len(lmlst)==21:
                    finger_count=[]
                    # all finger except thumb
                    for i in range(1,5): #i=1
                        if lmlst[tip_id[i]][2]>lmlst[tip_id[i]-2][2]:
                            finger_count.append(0)
                        else:
                            finger_count.append(1)
                    # print(finger_count)
                # print(id,lm)
            mp_drawings.draw_landmarks(image,x,mp_hands.HAND_CONNECTIONS)
        print(lmlst)
    cv2.imshow('Image',image)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
print(sum(finger_count))
```
